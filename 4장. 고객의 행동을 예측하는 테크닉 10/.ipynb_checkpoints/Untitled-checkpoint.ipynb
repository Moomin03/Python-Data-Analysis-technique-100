```python
# ÌÖåÌÅ¨Îãâ 31. Îç∞Ïù¥ÌÑ∞Î•º ÏùΩÏñ¥ Îì§Ïù¥Í≥† ÌôïÏù∏ÌïòÏûê
import pandas as pd
uselog = pd.read_csv('use_log.csv', parse_dates=['usedate'])
uselog.isnull().sum()
```

```python
customer = pd.read_csv('customer_join.csv')
customer.isnull().sum()
```

```python
# ÌÖåÌÅ¨Îãâ 32. ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏúº ÌöåÏõêÏùÑ Í∑∏Î£πÌôîÌïòÏûê
customer_clustering = customer[['mean', 'median', 'max', 'min', 'membership_period']]
customer_clustering.head()
```

```python
# ÌÖåÌÅ¨Îãâ 33. ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ Í≤∞Í≥ºÎ•º Î∂ÑÏÑùÌïòÏûê
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
customer_clustering_ss = ss.fit_transform(customer_clustering)
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
clusters = kmeans.fit(customer_clustering_ss)
customer_clustering['cluster'] = clusters.labels_
print(customer_clustering['cluster'].unique())
customer_clustering.head()
```

```python
customer_clustering.groupby('cluster').count()
```

```python
customer_clustering.groupby('cluster').mean()
```

```python
# ÌÖåÌÅ¨Îãâ 34. ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ Í≤∞Í≥ºÎ•º Í∞ÄÏãúÌôîÌïòÏûê
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(customer_clustering_ss)
result_pca = pca.transform(customer_clustering_ss)
pca_df = pd.DataFrame(result_pca)
pca_df['cluster'] = customer_clustering['cluster']
pca_df
```

```python
import matplotlib.pyplot as plt
pca_df.groupby('cluster').count()[0].plot(kind='pie')
```

```python
for i in customer_clustering['cluster'].unique():
    tmp = pca_df.loc[pca_df['cluster']==i]
    plt.scatter(tmp[0], tmp[1])
    plt.axis('off')
```

```python
# ÌÖåÌÅ¨Îãâ 45. ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ Í≤∞Í≥ºÎ•º Î∞îÌÉïÏúºÎ°ú ÌÉàÌá¥ ÌöåÏõêÏùò Í≤ΩÌñ•ÏùÑ ÌååÏïÖÌïòÏûê
concat_data = pd.concat(objs=[customer, customer_clustering['cluster']], axis='columns')
```

```python
# ÌÖåÌÅ¨Îãâ 36. Îã§Ïùå Îã¨Ïùò Ïù¥Ïö© ÌöüÏàò ÏòàÏ∏°ÏùÑ ÏúÑÌï¥ Îç∞Ïù¥ÌÑ∞Î•º Ï§ÄÎπÑÌïòÏûê
uselog['month'] = uselog['usedate'].dt.strftime('%Y%m')
uselog.head()
```

```python
uselog_month = uselog.groupby(by=['month', 'customer_id'], as_index=False).count()[['customer_id', 'month', 'log_id']]
uselog_month.rename(columns={'log_id':'count'}, inplace=True)
uselog_month.head()
```

```python
month_range = uselog_month['month'].unique()
month_range
```

```python
# ÌÖåÌÅ¨Îãâ 37, ÌäπÏßïÏù¥ ÎêòÎäî Î≥ÄÏàòÎ•º Ï∂îÍ∞ÄÌïòÏûê
plus_data = uselog_month.pivot_table(columns='month',
                         index='customer_id',
                         values='count')[month_range[5::-1]]
plus_data
```

```python
head_data = uselog_month[uselog_month['month'].between('201810', '201903')]
head_data
```

```python
head_data = pd.merge(how='left',
                     left=head_data,
                     right=plus_data,
                     on='customer_id')
head_data
```

```python
head_data.dropna(inplace=True)
head_data.reset_index(inplace=True, drop=True)
head_data.rename(columns={'count':'count_pred'}, inplace=True)
head_data
```

```python
head_data = pd.merge(how='left',
                     left=head_data,
                     right=customer[['customer_id', 'start_date']],
                     on='customer_id')
head_data
```

```python
head_data['start_date'] = pd.to_datetime(head_data['start_date'])
head_data['now_date'] = pd.to_datetime(head_data['month'], format="%Y%m")
head_data
```

```python
from dateutil.relativedelta import relativedelta
head_data['period'] = 0
for i in range(len(head_data)):
    delta = relativedelta(head_data.loc[i, 'now_date'], head_data.loc[i, 'start_date'])
    head_data.loc[i, 'period'] = delta.years*12+delta.months
head_data
```

```python
# ÌÖåÌÅ¨Îãâ 38. Îã§Ïùå Îã¨ Ïù¥Ïö© ÌöüÏàòÎ•º ÏòàÏ∏°ÌïòÎäî Î™®Îç∏ÏùÑ Íµ¨Ï∂ïÌïòÏûê
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
predict_data = head_data.loc[head_data['start_date']>=pd.to_datetime('20180401')]
predict_data
```

```python
lr = LinearRegression()
train_data = predict_data.drop(columns=['customer_id', 'month', 'count_pred', 'start_date', 'now_date'])
train_target = predict_data['count_pred']
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(train_data,
                                                                      train_target,
                                                                      test_size=0.2)
pipeline = make_pipeline(lr)
```

```python
pipeline.fit(train_input, train_target)
```

```python
pipeline.score(train_input, train_target)
```

```python
pipeline.score(test_input, test_target)
```

```python
# ÌÖåÌÅ¨Îãâ 39. Î™®Îç∏Ïóê Í∏∞Ïó¨ÌïòÎäî Î≥ÄÏàòÎ•º ÌôïÏù∏ÌïòÏûê
pd.DataFrame({
    'Features':pipeline.feature_names_in_,
    'Importances':pipeline.named_steps['linearregression'].coef_
}).sort_values(by='Importances', ascending=False)
```

```python
# ÌÖåÌÅ¨Îãâ 40. Îã§ÏùåÎã¨Ïùò Ïù¥Ïö© ÌöüÏàòÎ•º ÏòàÏ∏°ÌïòÏûê
x1 = [3, 4, 4, 6, 8, 7, 8]
x2 = [2, 2, 3, 3, 4, 6, 8]
x_pred = [x1, x2]
```

```python
pipeline.predict(x_pred)
```
