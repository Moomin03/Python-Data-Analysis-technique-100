```python
# í…Œí¬ë‹‰ 11. ë°ì´í„°ë¥¼ ì½ì–´ë“¤ì´ì
import pandas as pd
uriage_data = pd.read_csv('uriage.csv', parse_dates=['purchase_date'])
kokyaku_data = pd.read_excel('kokyaku_daicho.xlsx', parse_dates=['ë“±ë¡ì¼'])
```

```python
# í…Œí¬ë‹‰ 12. ë°ì´í„°ì˜ ì˜¤ë¥˜ë¥¼ ì‚´í´ë³´ì
uriage_data['item_name'].head()
```

```python
uriage_data.head()
```

```python
# í…Œí¬ë‹‰ 13. ë°ì´í„° ì˜¤ë¥˜ê°€ ìˆëŠ” ìƒíƒœë¡œ ì§‘ê³„ë¥¼ í•´ë³´ì
uriage_data['purchase_month'] = uriage_data['purchase_date'].dt.strftime('%Y%m')
uriage_data.head()
```

```python
res = uriage_data.pivot_table(columns='item_name',
                              index='purchase_month',
                              aggfunc='count',
                              fill_value=0)
res
```

```python
res = uriage_data.pivot_table(columns='item_name',
                              index='purchase_month',
                              values=['item_price'],
                              aggfunc='sum',
                              fill_value=0)
res
```

```python
# í…Œí¬ë‹‰ 14. ìƒí’ˆëª… ì˜¤ë¥˜ë¥¼ ìˆ˜ì •í•˜ì
uriage_data['item_name'] = uriage_data['item_name'].str.upper()
uriage_data['item_name'] = uriage_data['item_name'].str.replace(' ', '')

import numpy as np
np.unique(uriage_data['item_name'], return_counts=True)
```

```python
uriage_data.sort_values(by=['item_name'], inplace=True)
uriage_data
```

```python
# í…Œí¬ë‹‰ 15. ê¸ˆì•¡ì˜ ê²°ì¸¡ì¹˜ë¥¼ ìˆ˜ì •í•˜ì
uriage_data.isnull().sum()
```

```python
extracted_data = uriage_data[~(uriage_data['item_price'].isnull())].groupby(by='item_name').mean(numeric_only=True)
extracted_data.head()
```

```python
uriage_data.drop(columns='item_price',
                 axis='columns',
                 inplace=True)
uriage_data
```

```python
uriage_data = pd.merge(how='left',
                       left=uriage_data,
                       right=extracted_data,
                       on='item_name')
uriage_data.head()
```

```python
uriage_data.isnull().sum()
```

```python
np.unique(uriage_data['item_price'], return_counts=True)[1]
```

```python
np.unique(uriage_data['item_name'], return_counts=True)[1]
```

```python
# í…Œí¬ë‹‰ 16. ê³ ê° ì´ë¦„ì˜ ì˜¤ë¥˜ë¥¼ ìˆ˜ì •í•˜ì
kokyaku_data.head()
```

```python
kokyaku_data['ê³ ê°ì´ë¦„'] = kokyaku_data['ê³ ê°ì´ë¦„'].str.replace(' ', '')
kokyaku_data['ê³ ê°ì´ë¦„'] = kokyaku_data['ê³ ê°ì´ë¦„'].str.strip()
kokyaku_data.head()
```

```python
# í…Œí¬ë‹‰ 17. ë‚ ì§œ ì˜¤ë¥˜ë¥¼ ìˆ˜ì •í•˜ì
from_series = pd.to_timedelta(kokyaku_data[kokyaku_data['ë“±ë¡ì¼'].str.isdigit()]['ë“±ë¡ì¼'].astype(float), 
                              unit='days')+pd.to_datetime('1900/01/01')
from_series
```

```python
from_string = kokyaku_data[~(kokyaku_data['ë“±ë¡ì¼'].str.isdigit())]['ë“±ë¡ì¼']
from_string
```

```python
kokyaku_data['ë“±ë¡ì¼'] =  pd.concat(objs=[from_string, from_series])
kokyaku_data
```

```python
# í…Œí¬ë‹‰ 18. ê³ ê° ì´ë¦„ì„ í‚¤ë¡œ ë‘ê°œì˜ ë°ì´í„°ë¥¼ ê²°í•©í•˜ì
merged_data = pd.merge(how='left',
                       left=uriage_data,
                       right=kokyaku_data,
                       left_on='customer_name',
                       right_on='ê³ ê°ì´ë¦„')
merged_data
```

```python
merged_data.drop(columns=['ê³ ê°ì´ë¦„'],
                 axis='columns',
                 inplace=True)
merged_data
```

```python
# í…Œí¬ë‹‰ 19. ì •ì œí•œ ë°ì´í„°ë¥¼ ë¤í”„í•˜ì
dump_data = merged_data[['purchase_date', 'purchase_month', 'item_name', 'item_price', 'customer_name', 'ì§€ì—­', 'ë“±ë¡ì¼']]
dump_data
```

```python
dump_data.to_csv('dump.csv', index=False)
```
