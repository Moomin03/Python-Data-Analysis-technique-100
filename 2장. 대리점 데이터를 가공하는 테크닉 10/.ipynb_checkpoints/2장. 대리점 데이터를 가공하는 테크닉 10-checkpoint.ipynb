### 🗂️ 데이터 구성
---
    1. uriage.csv : 매출 이력
    2. kokyaku_daicho.xlsx : 대리점에서 관리하는 고객 정보

```python
# 테크닉 11. 데이터를 읽어들이자
import pandas as pd
uriage_data = pd.read_csv('uriage.csv', parse_dates=['purchase_date'])
kokyaku_data = pd.read_excel('kokyaku_daicho.xlsx', parse_dates=['등록일'])
```

```python
# 테크닉 12. 데이터의 오류를 살펴보자
uriage_data['item_name'].head()
```

```python
uriage_data.head()
```

```python
# 테크닉 13. 데이터 오류가 있는 상태로 집계를 해보자
uriage_data['purchase_month'] = uriage_data['purchase_date'].dt.strftime('%Y%m')
uriage_data.head()
```

```python
res = uriage_data.pivot_table(columns='item_name',
                              index='purchase_month',
                              aggfunc='count',
                              fill_value=0)
res
```

```python
res = uriage_data.pivot_table(columns='item_name',
                              index='purchase_month',
                              values=['item_price'],
                              aggfunc='sum',
                              fill_value=0)
res
```

```python
# 테크닉 14. 상품명 오류를 수정하자
uriage_data['item_name'] = uriage_data['item_name'].str.upper()
uriage_data['item_name'] = uriage_data['item_name'].str.replace(' ', '')

import numpy as np
np.unique(uriage_data['item_name'], return_counts=True)
```

```python
uriage_data.sort_values(by=['item_name'], inplace=True)
uriage_data
```

```python
# 테크닉 15. 금액의 결측치를 수정하자
uriage_data.isnull().sum()
```

```python
extracted_data = uriage_data[~(uriage_data['item_price'].isnull())].groupby(by='item_name').mean(numeric_only=True)
extracted_data.head()
```

```python
uriage_data.drop(columns='item_price',
                 axis='columns',
                 inplace=True)
uriage_data
```

```python
uriage_data = pd.merge(how='left',
                       left=uriage_data,
                       right=extracted_data,
                       on='item_name')
uriage_data.head()
```

```python
uriage_data.isnull().sum()
```

```python
np.unique(uriage_data['item_price'], return_counts=True)[1]
```

```python
np.unique(uriage_data['item_name'], return_counts=True)[1]
```

```python
# 테크닉 16. 고객 이름의 오류를 수정하자
kokyaku_data.head()
```

```python
kokyaku_data['고객이름'] = kokyaku_data['고객이름'].str.replace(' ', '')
kokyaku_data['고객이름'] = kokyaku_data['고객이름'].str.strip()
kokyaku_data.head()
```

```python
# 테크닉 17. 날짜 오류를 수정하자
from_series = pd.to_timedelta(kokyaku_data[kokyaku_data['등록일'].str.isdigit()]['등록일'].astype(float), 
                              unit='days')+pd.to_datetime('1900/01/01')
from_series
```

```python
from_string = kokyaku_data[~(kokyaku_data['등록일'].str.isdigit())]['등록일']
from_string
```

```python
kokyaku_data['등록일'] =  pd.concat(objs=[from_string, from_series])
kokyaku_data
```

```python
# 테크닉 18. 고객 이름을 키로 두개의 데이터를 결합하자
merged_data = pd.merge(how='left',
                       left=uriage_data,
                       right=kokyaku_data,
                       left_on='customer_name',
                       right_on='고객이름')
merged_data
```

```python
merged_data.drop(columns=['고객이름'],
                 axis='columns',
                 inplace=True)
merged_data
```

```python
# 테크닉 19. 정제한 데이터를 덤프하자
dump_data = merged_data[['purchase_date', 'purchase_month', 'item_name', 'item_price', 'customer_name', '지역', '등록일']]
dump_data
```

```python
dump_data.to_csv('dump.csv', index=False)
```

```python
# 테크닉 20. 데이터를 집계하자
dump_data.pivot_table(index='purchase_month',
                      columns='item_name',
                      aggfunc='count',
                      fill_value=0)
```
